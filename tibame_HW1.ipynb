````python
import random
# 凱薩密碼的加密
s = "AppleZ"
# 以前: lower(s)
# 專屬功能: s.lower()
s_lower = s.lower()
# 位移幾個
password = random.randint(1, 25)

result = ""
for c in s_lower:
    # 取得ASCII->位移到0-25
    n = ord(c) - ord("a")
    # 往後位移做出密碼
    n = (n + password) % 26 + ord("a")
    # 放到result
    result = result + chr(n)
print(password, result)
````
