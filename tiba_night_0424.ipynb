````python
import random

my = int(input("0.剪刀 1.石頭 2.布:"))
com = random.randint(0, 2)

trans = ["剪刀", "石頭", "布"]
print("我出的", trans[my])
print("電腦的", trans[com])

if my == (com + 1) % 3:
    print("WIN")
elif com == (my + 1) % 3:
    print("LOSE")
else:
    print("EVEN")
````

````python
import random

# my = int(input("0-5:"))
# com = random.randint(0, 5)
# 0 3 even
# 0 2 win
# 3 5 win
my = 0
com = 2

trans = ["剪刀(0)", "石頭(0)", "布(0)", "剪刀(1)", "石頭(1)", "布(1)"]
print("我出的", trans[my])
print("電腦的", trans[com])

# my_g: 0 or 1
my_g = my // 3
com_g = com // 3

if my_g == com_g:
    print("同組")
    # case: 0 1 2
    # case: 3 4 5 -> 0 1 2
    my2 = my % 3
    com2 = com % 3
    # print(my2, com2)
    if my2 == (com2 + 1) % 3:
        print("WIN")
    elif com2 == (my2 + 1) % 3:
        print("LOSE")
    else:
        print("EVEN")
else:
    print("不同組")
````
